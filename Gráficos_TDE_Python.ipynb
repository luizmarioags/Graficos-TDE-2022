## Variações Percentuais no PIB per Capita de 5 em 5 anos

## Para calcularmos essas variações usaremos a seguinte expressão:
\begin{equation}
\frac{PIB_{Ano Corrente} - PIB_{Ano Base}}{PIB_{Ano Base}}
\end{equation}

```python
# Calculando as variações percentuais no PIB per Capita do Brasil 
# Primeiro iremos construir séries para o PIB do ano Corrente 
PIBperCapita_Ano_Corrente = data_set_Brasil['PIB per Capita']
PIBperCapita_Ano_Corrente
```

```python
# Construindo a série para o PIB do ano Base
PIBperCapita_Ano_Base = PIBperCapita_Ano_Corrente.shift(1)
PIBperCapita_Ano_Base
```

```python
# Calculando as variações percentuais do PIB per Capita 
data_set_Brasil["Variação Percentual do PIB per Capita"] = ((PIBperCapita_Ano_Corrente/PIBperCapita_Ano_Base) - 1)*100
data_set_Brasil
```

```python
# Plotando o gráfico com o ploty 
# Gráfico Interativo no Ploty 
fig = px.line(data_set_Brasil, x="YearCode", y="Variação Percentual do PIB per Capita", title='Variação Percentual do PIB per Capita - Brasil (1950 - 2019) com intervalo de 5 em 5 anos')
fig.show()
```

```python
# Gráfico com o Seaborn 
sns.set()
sns.set(rc={"figure.figsize":(8, 4)})
sns.lineplot(data = data_set_Brasil, x ='YearCode', y ='Variação Percentual do PIB per Capita', color = 'blue')
plt.title('Variação Percentual do PIB per Capita - Brasil (1950 - 2019), de 5 em 5 anos.\nFeito por Luiz Mario. Fonte: Penn World Table', loc='center')
```

```python
# Calculando as Variações Percentuais do PIB per capita da Coreia do Sul 
# Definindo o PIB per Capita no Ano Corrente - Coreia do Sul
PIBperCapita_Ano_Corrente_Coreia = data_set_Coreia_Sul['PIB per Capita']
PIBperCapita_Ano_Corrente_Coreia
```

```python
# Definindo o PIB per Capita no Ano Base - Coreia do Sul 
PIBperCapita_Ano_Base_Coreia = PIBperCapita_Ano_Corrente_Coreia.shift(1)
PIBperCapita_Ano_Base_Coreia
```

```python
# Calculando as variações percentuais do PIB per Capita para a Coreia do Sul
data_set_Coreia_Sul["Variação Percentual do PIB per Capita"] = ((PIBperCapita_Ano_Corrente_Coreia/PIBperCapita_Ano_Base_Coreia) - 1)*100
data_set_Coreia_Sul
```

```python
# Gráfico Interativo no Ploty 
fig = px.line(data_set_Coreia_Sul, x="YearCode", y="Variação Percentual do PIB per Capita", title='Variação Percentual do PIB per Capita - Coreia do Sul (1955 - 2019) com intervalo de 5 em 5 anos')
fig.show()
```

## Tendo como base a seguinte função de produção do tipo Cobb-Douglas: $Y(K(t),L(t)) = A(t)K(t)^αL(t)^{1 - α}$  
Em que $Y$ é o produto, $K(t)$ é o capital, $L(t)$ é o trabalho, $\alpha$ é representa a fração do produto que vai para o capital ($K$) e para o trabalho ($L$) e por fim o coeficiente $A(t)$ representa a produtividade total dos fatores, um aumento nele leva a um aumento do produto para a mesma quantidade de capital e trabalho. Desde que suas produtividades marginais sejam positivas e constantes, isto é Ceteris Paribus. 

## Iremos agora brincar com o coeficiente $A(t)$, iremos calcular ele para diversos países e realizar os gráficos em relação ao tempo!!!!. Porém primeiremnte é necessário dizer que: 


*  Juros ou Remuneração do Capital: $r = \dfrac{\partial Y(K,L)}{\partial K} = \alpha A(t)K^{\alpha - 1}L^{1 - \alpha}$ 
*   Salários: $w = \dfrac{\partial Y(K,L)}{\partial L} = (1 - \alpha) A(t)K^{\alpha}L^{-α}$


Logo podemos definir o seguinte:
* Renda do Trabalho: $wL = (1 - \alpha)Y → \frac{wl}{Y} = 1 - \alpha$
* Renda do Capital: $rK = \alpha Y → \alpha = \frac{rk}{Y}$



## Com essas definições postas vamos ao código e aos gráficos
