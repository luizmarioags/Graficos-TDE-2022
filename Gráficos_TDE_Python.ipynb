```python
# Importando bibliotecas necessárias
import pandas as pd 
import numpy as np 
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
# Criando data sets ou data frames para cada país
data_set_Brasil = pd.read_excel('PIB e População - Brasil (1950-2019).xlsx')
data_set_Coreia_Sul = pd.read_excel('PIB e População - Coreia do Sul (1950 - 2019).xlsx')
data_set_Estados_Unidos = pd.read_excel('PIB e População - Estados Unidos (1950 - 2019).xlsx')
```

```python
# Vizualizando o data set do Brasil
data_set_Brasil
```

```python
# Vizualizando o data set da Coreia do Sul
data_set_Coreia_Sul
```

```python
# Vizualizando o data set dos Estados Unidos
data_set_Estados_Unidos
```

```python
# Calculando o PIB per Capita
PIB_Per_Capita_Brasil = data_set_Brasil['RGDPO']/data_set_Brasil['Population']
PIB_Per_Capita_Brasil
```

```python
# Criando o array numpy 
PIB_Per_Capita_Brasil = np.array(PIB_Per_Capita_Brasil)
# Adicionado o array no data set
data_set_Brasil['PIB per Capita'] = pd.Series(PIB_Per_Capita_Brasil)
data_set_Brasil
```

```python
# Teste Ploty 
fig = px.line(data_set_Brasil, x="YearCode", y="PIB per Capita", title='PIB per Capita - Brasil (1950 - 2019)')
fig.show()
```

```python
# Plotando com o Seaborn
sns.set()
sns.lineplot(data = data_set_Brasil, x ='YearCode', y ='PIB per Capita')
```

## Agora para a Coreia do Sul

```python
# Calculando o PIB per Capita da Coreia do Sul 
PIB_Per_Capita_Coreia = data_set_Coreia_Sul['RGDPO']/data_set_Coreia_Sul['Population']
PIB_Per_Capita_Coreia
```

```python
# Criando o array numpy 
PIB_Per_Capita_Coreia = np.array(PIB_Per_Capita_Coreia)
# Adicionando o Array no data set
data_set_Coreia_Sul['PIB per Capita'] = pd.Series(PIB_Per_Capita_Coreia)
data_set_Coreia_Sul
```

```python
# Gráfico Interativo no Ploty 
fig = px.line(data_set_Coreia_Sul, x="YearCode", y="PIB per Capita", title='PIB per Capita - Coreia do Sul (1955 - 2019)')
fig.show()
```

```python
# Seaborn
sns.set()
sns.lineplot(data = data_set_Coreia_Sul, x ='YearCode', y ='PIB per Capita', color = 'red')
```

```python
# Plotando o gráfico comparativo dos dois países com o Ploty 
# Carregando os data sets necessários 
# Data set do Brasil 
anos = data_set_Brasil['YearCode']
PIB_Per_Capita_Brasil_Plot = data_set_Brasil['PIB per Capita']
# Data set da Coreia 
anos_coreia = data_set_Coreia_Sul['YearCode']
PIB_Per_Capita_Coreia_Plot = data_set_Coreia_Sul['PIB per Capita']
# Plotando as duas Curvas com o Ploty para fins comparativos 
fig = go.Figure()
# Adicionando as curvas 
fig.add_trace(go.Scatter(x = anos , y = PIB_Per_Capita_Brasil_Plot , name = 'Brasil', line = dict(color = 'royalblue', width = 4)))
fig.add_trace(go.Scatter(x = anos_coreia , y = PIB_Per_Capita_Coreia_Plot, name = 'Coreia do Sul', line = dict(color = 'firebrick', width = 4)))
# Editando o layout do gráfico 
fig.update_layout(title='PIB per Capita comparado - Brasil e Coreia do Sul (1950 - 2019)',
                   xaxis_title='Anos',
                   yaxis_title='PIB per Capita')
fig.show()
```

```python
# Plotando as duas curvas com o Seaborn 
sns.set()
sns.lineplot(data = data_set_Brasil, x ='YearCode', y ='PIB per Capita', color = 'blue')
sns.lineplot(data = data_set_Coreia_Sul, x ='YearCode', y ='PIB per Capita', color = 'red')
plt.legend(labels = ["Brasil","Coreia do Sul"])
```

```python
# Matplotlib
ns.lineplot(data = data_set_Brasil, x ='YearCode', y ='PIB per Capita', color = 'blue')
sns.lineplot(data = data_set_Coreia_Sul, x ='YearCode', y ='PIB per Capita', color = 'red')
plt.legend(labels = ["Brasil","Coreia do Sul"])
```

## Variações Percentuais no PIB per Capita

```python
# Calculando as variações percentuais no PIB per Capita do Brasil 
# Primeiro iremos construir séries para o PIB do ano Corrente 
PIBperCapita_Ano_Corrente = data_set_Brasil['PIB per Capita']
PIBperCapita_Ano_Corrente
```

```python
# Construindo a série para o PIB do ano Base
PIBperCapita_Ano_Base = PIBperCapita_Ano_Corrente.shift(1)
PIBperCapita_Ano_Base
```

```python
# Calculando as variações percentuais do PIB per Capita 
data_set_Brasil["Variação Percentual do PIB per Capita"] = ((PIBperCapita_Ano_Corrente/PIBperCapita_Ano_Base) - 1)*100
data_set_Brasil
```

```python
# Plotando o gráfico com o ploty 
# Gráfico Interativo no Ploty 
fig = px.line(data_set_Brasil, x="YearCode", y="Variação Percentual do PIB per Capita", title='Variação Percentual do PIB per Capita - Brasil (1950 - 2019) com intervalo de 5 em 5 anos')
fig.show()
```

```python
# Gráfico com o Seaborn 
sns.set()
sns.set(rc={"figure.figsize":(8, 4)})
sns.lineplot(data = data_set_Brasil, x ='YearCode', y ='Variação Percentual do PIB per Capita', color = 'blue').set(title = 'Variação Percentual do PIB per Capita - Brasil (1950 - 2019), de 5 em 5 anos.')
```

```python
# Calculando as Variações Percentuais do PIB per capita da Coreia do Sul 
# Definindo o PIB per Capita no Ano Corrente - Coreia do Sul
PIBperCapita_Ano_Corrente_Coreia = data_set_Coreia_Sul['PIB per Capita']
PIBperCapita_Ano_Corrente_Coreia
```

```python
# Definindo o PIB per Capita no Ano Base - Coreia do Sul 
PIBperCapita_Ano_Base_Coreia = PIBperCapita_Ano_Corrente_Coreia.shift(1)
PIBperCapita_Ano_Base_Coreia
```

```python
# Calculando as variações percentuais do PIB per Capita para a Coreia do Sul
data_set_Coreia_Sul["Variação Percentual do PIB per Capita"] = ((PIBperCapita_Ano_Corrente_Coreia/PIBperCapita_Ano_Base_Coreia) - 1)*100
data_set_Coreia_Sul
```

```python
# Gráfico Interativo no Ploty 
fig = px.line(data_set_Coreia_Sul, x="YearCode", y="Variação Percentual do PIB per Capita", title='Variação Percentual do PIB per Capita - Coreia do Sul (1955 - 2019) com intervalo de 5 em 5 anos')
fig.show()
```

```python
# Gráfico no Seaborn 
sns.set()
sns.set(rc={"figure.figsize":(8, 4)})
sns.lineplot(data = data_set_Coreia_Sul, x ='YearCode', y ='Variação Percentual do PIB per Capita', color = 'red').set(title = 'Variação Percentual do PIB per Capita - Coreia do Sul (1955 - 2019), de 5 em 5 anos.')
```

```python
# Plotando as duas curvas com o ploty 
# Carregando os data sets necessários 
# Data set do Brasil 
anos = data_set_Brasil['YearCode']
PIB_Per_Capita_Brasil_Plot = data_set_Brasil['Variação Percentual do PIB per Capita']
# Data set da Coreia 
anos_coreia = data_set_Coreia_Sul['YearCode']
PIB_Per_Capita_Coreia_Plot = data_set_Coreia_Sul['Variação Percentual do PIB per Capita']
# Plotando as duas Curvas com o Ploty para fins comparativos 
fig = go.Figure()
# Adicionando as curvas 
fig.add_trace(go.Scatter(x = anos , y = PIB_Per_Capita_Brasil_Plot , name = 'Brasil', line = dict(color = 'royalblue', width = 4)))
fig.add_trace(go.Scatter(x = anos_coreia , y = PIB_Per_Capita_Coreia_Plot, name = 'Coreia do Sul', line = dict(color = 'firebrick', width = 4)))
# Editando o layout do gráfico 
fig.update_layout(title='Variação Percentual do PIB per Capita Comparado - Brasil e Coreia do Sul (1950 - 2019)',
                   xaxis_title='Anos',
                   yaxis_title='Variação Percentual do PIB per Capita')
fig.show()
```

```python
# Plotando as duas curvas com o Seaborn 
sns.set()
sns.set(rc={"figure.figsize":(8, 4)})
sns.lineplot(data = data_set_Brasil, x ='YearCode', y ='Variação Percentual do PIB per Capita', color = 'blue')
sns.lineplot(data = data_set_Coreia_Sul, x ='YearCode', y ='Variação Percentual do PIB per Capita', color = 'red')
plt.legend(labels = ["Brasil","Coreia do Sul"])
```
